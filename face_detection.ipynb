```
# Carregando Dependências para processamento de imagens, análise de dados e etc
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
import scipy
```

```
# Atribuído imagem à variável urls
urls = ["https://1734811051.rsc.cdn77.org/data/images/full/381803/deepfake-images.png"]

# Verificando e mostrando a imagem padrão
for url in urls:
  image = io.imread(url) 
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  
cv2_imshow(image_2)
```

```
## Inserindo Retângulo na face

# técnica de visão computacional pra detectar faces humanas
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml') # arquivo específico para detectar frontal face

# Definindo parâmetro de escala, tamanho mínimo do objeto e eliminando falsos positivos
faces = face_cascade.detectMultiScale(image_2, scaleFactor = 1.04, minNeighbors = 7, minSize = (30,30), flags = cv.CASCADE_SCALE_IMAGE)

# Implementando cor ao retângulo e coordenadas:
for (x,y,w,h) in faces:
     cv.rectangle(image_2,(x,y),(x+w,y+h),(0,191,255),2)

cv2_imshow(image_2)
```

```
## Inserindo Círculo na face
for (x, y, w, h) in faces:
    coordinates = x + w // 2, y + h // 2 # coordenadas do ponto médio
    radius = w // 2 # raio do círculo
    cv.circle(image_2, coordinates, radius, (255, 0, 255), 3) # definido a cor e atribuído o circle

cv2_imshow(image_2)
```
